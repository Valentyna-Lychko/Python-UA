Таблиця- кольрів  для проету makeup

| Назва колонки або категорії | HEX-код    | Назва українською         |
|-----------------------------|------------|---------------------------|
| name                       | `#80b1d3`  | Світло-синій              |
| brand_category_product_type | `#d2a5d0`  | Яскравий рожево-фіолетовий |
| category_product_type       | `#58c4a3`  | Яскравий бірюзовий         |
| brand                      | `#bc80bd`  | Фіолетовий                |
| category                   | `#b3de69`  | Світло-зелений                |
| product_type               | `#ffed6f`  | Світло-золотий            |
| low (price category)       | `#fdbf6f`  | Світло-помаранчевий       |
| mid (price category)       | `#fb9a99`  | Світло-рожевий            |
| high (price category)      | `#ff7f00`  | Помаранчевий              |
| premium (price category)   | `#e31a1c`  | Яскраво-червоний          |
| rating                     | `#4B0082`  | Індиго                    |
| price                      | `#8c654b`  | Темно-коричневий          |


```python
#створення таблиці кольрів

# Дані для таблиці
data = {
    "Column or Category Name": [
        "name", "brand_category_product_type", "category_product_type",
        "brand", "category", "product_type", "low (price category)",
        "mid (price category)", "high (price category)", "premium (price category)",
        "rating", "price"
    ],
    "HEX Code": [
        "#80b1d3", "#d2a5d0", "#58c4a3", "#bc80bd", "#b3de69", "#ffed6f",
        "#fdbf6f", "#fb9a99", "#ff7f00", "#e31a1c", "#4B0082", "#8c654b"
    ],
    "Color Name": [
        "Light Blue", "Bright Pink-Violet", "Bright Turquoise",
        "Violet", "Light Green", "Light Gold", "Light Orange",
        "Light Pink", "Orange", "Bright Red", "Indigo", "Dark Brown"
    ]
}

# Створення DataFrame
df = pd.DataFrame(data)

# Генерація HTML таблиці з кольоровими блоками
html_table = """
<table>
    <tr>
        <th>Column or Category Name</th>
        <th>HEX Code</th>
        <th>Color Name</th>
        <th>Color</th>
    </tr>
"""

for _, row in df.iterrows():
    html_table += f"""
    <tr>
        <td>{row['Column or Category Name']}</td>
        <td>{row['HEX Code']}</td>
        <td>{row['Color Name']}</td>
        <td style="background-color:{row['HEX Code']}; width: 100px; height: 20px;"></td>
    </tr>
    """

html_table += "</table>"

# Збереження таблиці в HTML-файл
with open("color_table.html", "w", encoding="utf-8") as f:
    f.write(html_table)

print("Table with colors saved to 'color_table.html'")
```
